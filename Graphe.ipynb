```python
import matplotlib.pyplot as plt

# 测试的数据集大小
sizes = [1000, 5000, 10000, 20000, 50000, 80000, 120000, 200000]

# 对应的平均执行时间（微秒）
times = [48, 212, 423, 912, 1973, 3409, 5146, 8510]

plt.plot(sizes, times, marker='o')
plt.title('Ajouts Iteratifs tableau')
plt.xlabel('Size')
plt.ylabel('Execution Time (10^-6s)')
plt.grid(True)
plt.show()
```
