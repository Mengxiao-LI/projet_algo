```python
import matplotlib.pyplot as plt

# 测试的数据集大小
sizes = [1000, 5000, 10000, 20000, 50000, 80000, 120000, 200000]

# 对应的平均执行时间（微秒）
times = [48, 212, 423, 912, 1973, 3409, 5146, 8510]

plt.plot(sizes, times, marker='o')
plt.title('Ajouts Iteratifs tableau')
plt.xlabel('Size')
plt.ylabel('Execution Time (10^-6s)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 测试数据
methods = ['AjoutsIteratifs tas', 'AjoutFB FileBinomiale']
times = [664, 20714575]  # 微秒

# 绘制直方图
plt.figure(figsize=(10, 6))
plt.bar(methods, times, color=['blue', 'green'])
plt.ylabel('Time (microseconds)')
plt.title('Comparison of Execution Time')
plt.yscale('log')  # 使用对数刻度，因为时间差异很大
plt.show()
```
