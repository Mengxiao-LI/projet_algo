```python
import pandas as pd
import matplotlib.pyplot as plt

#  CSV 
df = pd.read_csv('construction_times2.csv')

# 
plt.figure(figsize=(10, 6))
plt.plot(df['Count'], df['AverageTime'], marker='o')
plt.title('Construction Time vs Number of Keys')
plt.xlabel('Number of Keys')
plt.ylabel('Time (seconds)')
plt.grid(True)
plt.show()
```
